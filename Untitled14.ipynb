### plot of the partial sum

```python
def sum(N,x):
    S = 0
    for n in range(1,N+1):
        S+=1/(2*n-1)**2*np.cos((2*n-1)*x)
    f_N = (np.pi/2)-(4/np.pi)*S
    return f_N
```

```python
x = np.linspace(-np.pi, np.pi, 100)
f_5 = sum(5,x)
f_15 = sum(15,x)
f_30 = sum(30,x)
```

```python
plt.plot(x,f_5,'--', lw=5);
plt.plot(x,f_15,'-');
plt.plot(x,f_30,lw=2);
```
